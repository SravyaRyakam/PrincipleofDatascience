```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Loading the cleaned data
cleaned_data = pd.read_csv("/content/Preprocess_data1.csv")
print(cleaned_data.describe())

```

```python
# Visualization 1: Histogram of weight
plt.figure(figsize=(8, 6))
sns.histplot(cleaned_data['Weight'], bins=10, kde=True)
plt.title("Histogram of Weight")
plt.savefig("visualizations of Weight.png")
plt.show()
```

```python
# Visualization2: Scatter plot of Grip strength vs Weight
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Grip strength (kg)', data=cleaned_data)
plt.title("Scatter plot of Gripstrength vs Weight")
plt.savefig("visualizations of scatter_Weight_grip.png")
plt.show()
```
