```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/student.csv")
```

```python
# Visualization 1: Gender Distribution
gender_counts = data['gender'].value_counts()
plt.figure(figsize=(6, 4))
gender_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.savefig("visualizations of Gender Distribution.png")
```

```python
# Visualization : Lunch Type
lunch_counts = data['lunch'].value_counts()
plt.figure(figsize=(6, 6))
lunch_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Lunch Type Distribution')
plt.savefig("visualizations of Luch Type Distribution.png")
```

```python
# Visualization : Math Score Distribution
plt.figure(figsize=(8, 4))
data['math score'].plot(kind='hist', bins=20, color='skyblue')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('Math Score Distribution')
plt.savefig("visualizations of Math Score Distribution.png")
```

```python
# Visualization : Reading Score Distribution
plt.figure(figsize=(8, 4))
data['reading score'].plot(kind='hist', bins=20, color='salmon')
plt.xlabel('Reading Score')
plt.ylabel('Frequency')
plt.title('Reading Score Distribution')
plt.savefig("visualizations of Reading Score Distribution.png")
```

```python
# Visualization : Scatter Plot of Math vs. Reading Scores
plt.figure(figsize=(8, 6))
plt.scatter(data['math score'], data['reading score'], alpha=0.5, color='purple')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.title('Scatter Plot: Math vs. Reading Scores')
plt.savefig("visualizations of Scatter Plot: Math vs. Reading Scores.png")
```

```python
# Visualization : Scatter Plot of Math vs. Reading Scores
plt.figure(figsize=(8, 6))
data[['math score', 'reading score', 'writing score']].boxplot()
plt.title('Box Plot of Scores')
plt.ylabel('Score')
plt.show()

plt.savefig("visualizations of BOX Plot: Math vs Reading Scores vs writing score.png")
```
